# Seaborn Exercises

1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt # not strictly necessary for most plotting applications
```

```python
# Load the "tips" dataset 
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 1. add a percentage tip column to the dataset
tips['percent_tip'] = tips['tip']/tips['total_bill']
tips.head()
```

```python
# 2. visualize the relationship between tip percentage, with day of week and time 
sns.catplot(x="day", y="percent_tip", hue="time", kind="swarm", data=tips)
```

```python
# 3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)
sns.pairplot(tips, x_vars = ["day", "time", "size"], y_vars = ["tip"])
```

2) Find an interesting trend in the "exercise" dataset

```python
# load the "exercise" dataset
exercise = sns.load_dataset('exercise')
exercise.head()
```

```python
# visualization of the data to see any intersting trends or clustering 
# time, pulse, diet
sns.pointplot(x='time',y='pulse',hue='diet',data=exercise)
plt.show()
print('No fat diet individuals tends to have higher pulses for all time comparing to low fat diet individuals, but as the exercise \ntime increased, the difference increases.')
```

```python
# kind, pulse, diet 
sns.catplot(x="kind", y = "pulse", hue = "diet", data=exercise)
plt.show()
print('No fat diet individuals has faster pulse in running conditions comparing to low fat diet individuals.')
```

```python
# interaction 
sns.catplot(x='diet',y='pulse',hue = 'kind',col = 'time',data=exercise,kind='point')
plt.show()
```
